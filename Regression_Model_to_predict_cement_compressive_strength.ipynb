Regression Model to predict Cement compressive strength

```python
import pandas as pd
```

```python
cement = pd.read_csv('http://github.com/ybifoundation/Dataset/raw/main/Concrete%20Compressive%20Strength.csv')
```

```python
cement.head()
```

```python
cement.nunique()
```

```python
import seaborn as sns
sns.pairplot(cement)
```

```python
cement.columns
```

```python
y=cement[cement.columns[-1]]
cement.shape
```

```python
X=cement.drop('Concrete Compressive Strength(MPa, megapascals) ',axis=1)
X.shape
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=.3,random_state=40)
```

```python
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
y_pred=model.predict(xtest)
```

```python
model.score(xtest,ytest)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
mean_absolute_error(y_true=ytest,y_pred=y_pred)
```

```python
import numpy as np
def MAPE(Y_actual,Y_Predicted):
  mape = np.mean(np.abs((Y_actual - Y_Predicted)/Y_actual))
  return mape
```

```python
MAPE(ytest,y_pred)
```

```python
mean_squared_error(y_true=ytest,y_pred=y_pred)
```
